```python
data.rename(columns={'sex':'gender'},inplace=True)
data.head()
```

```python
# Changing gender values to categories
data.loc[data['gender'] == 0, 'gender'] = 'female'
data.loc[data['gender'] == 1, 'gender'] = 'male'

# Check the updated DataFrame
data.head()
```

 Categorical: chest_pain_type,Sex,fasting_blood_sugar,rest_ecg,exercise_induced_angina,st_slope,thalassemia
 Numerical Features: Age (Continuous),resting_blood_pressure,cholesterol,max_heart_rate_achieved,st_depression,num_major_vessels,target

```python
#MORE OF THE FEMALES ARE HAVING HEART DISEASE
sns.barplot(data=data, x='gender', y='target')

# Add title
plt.title('Heart Disease Distribution by Gender')

# Calculate and print the percentage of females with heart diseases
female_percentage = data['target'][data['gender'] == 'female'].value_counts(normalize=True).get(1, 0) * 100
print('Percentage of females with Heart Diseases:', female_percentage)

# Calculate and print the percentage of males with heart diseases
male_percentage = data['target'][data['gender'] == 'male'].value_counts(normalize=True).get(1, 0) * 100
print('Percentage of males with Heart Diseases:', male_percentage)



```

```python
data[['gender','target']].groupby(['gender']).mean()
```

```python
f,ax = plt.subplots(1,2,figsize=(10,5))
data[['gender','target']].groupby(['gender']).mean().plot.bar(ax=ax[0])
ax[0].set_title("Target vs Gender")
sns.countplot(x='gender',hue= 'target',data=data,ax = ax[1])
ax[1].set_title("GENDERWISE HEALTHY vs HEART DISEASE")
##According to the graph most of the men are healthy and few have heart disease which is reverse in female's case
```

```python
f,ax = plt.subplots(1,2,figsize=(10,5))
data['cp'].value_counts().plot.bar(ax=ax[0])
sns.countplot(x='cp',hue='target',data=data,ax=ax[1])

## Mostly people who have 2nd type of chest pain i.e. Non-Angina are likely to have a heart disease ##
```

```python
sns.countplot(x='cp',hue='gender',data=data).set_title("GENDERWISE CHEST PAIN")
```

```python
f,ax = plt.subplots(1,2,figsize=(18,8))
sns.violinplot(x='fbs',y='age',hue='target',data=data,split=True,ax=ax[0]).set_title("TARGET-WISE FASTING BLOOD SUGAR")
sns.violinplot(x='fbs',y='age',hue='gender',data=data,split=True,ax=ax[1]).set_title("TARGET-WISE FASTING BLOOD SUGAR")
```

```python
sns.stripplot(x='gender',y='chol',hue='age',data=data,jitter=True,edgecolor='green')
```

```python
# Create a crosstab for 'target' and 'gender'
tab = pd.crosstab(data['target'], data['gender'])
print(tab)

# Normalize the crosstab table row-wise to get percentages
dummy = tab.div(tab.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True)

# Add labels to the plot
plt.xlabel('Target')
plt.ylabel('Percentage')

# Add a title to the plot
plt.title('Gender Distribution by Target Category')

#Show the graph
plt.show()

```

```python
# Scatter plot between 'age' and 'oldpeak'
plt.scatter(data['age'], data['oldpeak'], color='b', alpha=0.6)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Oldpeak')
plt.title('Scatter Plot of Age vs. Oldpeak')

# Show the plot
plt.show()
```

```python
# Convert gender to numerical values
data['gender'] = data['gender'].replace({'male': 1, 'female': 0})

# Calculate the correlation matrix again
corr_matrix = data.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap with Categorical Variables Encoded', fontsize=16)
plt.show()
```
